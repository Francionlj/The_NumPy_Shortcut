# 1.9 Transposicao de arrays e trocas de eixo

```python
import numpy as np
```

A trans posicao e uma forma especial de reformatacao que tenha tambem reorna uma visaualisacao dos dados subjacentes, sem fazer nenhuma copia. Os arrays tem o metodo trenspose e o atributo especial T

```python
arr = np.arange(15).reshape((3, 5))
```

```python
print(arr)
```

```python
print(arr.T)
```

Ao fazer calculo com matriz talvez voce use esse recurso com alguma frequancia por exemplo,  ao calcular da matriz inversa interna usando `numpy. dot`

```python
arr = np.array([[0, 1, 0], [1, 2, -2], [6, 3, 2], [-1, 0, -1], [1, 0 , 1]])
```

```python
print(np.dot(arr.T,  arr))
```

Opercao infixo `@` e outra maneira para fazermos a multiplicacao de matrizeis:

```python
print(arr.T @ arr)
```

A transposicao simpes com `.T` e um caso especifico de troca de eixos. O ndarraytem o metodo `swapaxes`,  que recebe um par de numeros de eixo e troca os eixos indicados para reorganisar os dados

```python
print(arr)
```

```python
print(arr.swapaxes(0, 1))
```

O metodo `swapaxes` tamnem retorna uma visualisacao dos dados sem fazer copia
