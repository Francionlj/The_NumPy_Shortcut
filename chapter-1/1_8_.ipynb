# 1.6 Indexacao booleana

```python
import numpy as np
```

Considerando um exemplo no qual temos alguns dados em um array e um array de nomes com duplicadas

```python
names = np.array(['Bob', 'Joe', 'Will', 'Bob',  'Joe', 'Will',  "Joe" ])
```

```python
names
```

```python
data = np.array([[4, 7], [0,2], [-5, 6], [0,0], [1, 2], [-12, -4], [3, 4]])
```

```python
print(data)
```

Suponhamos que cada nome corresponde uma linha de um array data e quisermos selecionar todas as linhas cujo nome correespomdemte foss `Bob`.  com as operacoes aritimedicas,  as comparacoes (como com ==) com arrays tambem sao vetorisaodos.  Logo comparar `names` com `string` "Bobâ€ gera um array bpleano

```python
names == 'Bob'
```

Esse array booleano pode ser passado quando o array for indexado:

```python
print(data[names == 'Bob'])
```

O array boleano deve ter o mesmo tamanho do eixo de array que ele esta indexado.  Voce pode ate mesmo combinar arrays booleanos com fatias ou inteiros (ou sequancias de inteiros)

Nestes exemplos,  estou fazendo a selecao nas linhas em que `names == Bob` e tambem estou indexando as colunas

```python
print(data[names == 'Bob',  1:])
```

```python
print(data[names == 'Bob',  1])
```

Para celcionar tudo exceto Bob voce pode usar `!=` ou nefar a condiccao utilisanmco `~`


```python
names != 'Bob'
```

```python
~(names == 'Bob')
```

```python
print(data[~(names == 'Bob')])
```

O operdos `~` pode ser util quando voce quiser inverter um array booleana referencido por uma variavel:

```python
cond = names == 'Bob'
```

```python
cond
```

```python
print(data[~cond])
```

Para selecionar dois dos tres nomes e combinar varias condicoes booleanas, use operadores aritimeticas booleanas `&` (and) e `|` (or)

```python
mask = (names == 'Bob') | (names == 'Will')
```

```python
mask
```

```python
print(data[mask])
```

A selecao de dados de um array por indexacao booleana e a atribuicao do resultado a uma nova variavel sempre cria uma copia dos dados,  mesmo se o array retornado nao for aleatorio.

Obs: apavars `and` e `or` do Python nao funciona com arrays boenaos.

A definicao de valores com arrays booleanos funciona pela trensferencia de valores ou dos valores que estiverem no lado dierito para os locais em que os valores do array booleano forem `True`.  Parta configurar todos os valores negativos de `data` com 0, so precisamos fazer o seguint:

```python
data[data < 0] = 0
```

```python
print(data)
```

```python
print(data[names == 'Bob',  1:])
```

Voce tambem pode definir linhas e colunas inteiras usando um array booleano unidimencional

```python
data[names != 'Joe'] = 7
```

```python
print(data)
```

Como veremos posteriormente,  esses tipos de opareacoes com dados bidimencionais sao mais convenientes com pandas>
