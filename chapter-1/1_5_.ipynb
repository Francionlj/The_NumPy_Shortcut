# 1.4 Aritimetica com o NUmpy

```python
import numpy as np
```

OS arrays sao importantes porque permitem expressar operacoes em lote como dados sem ser preciso escrevre nenhum lote `for`. Os usuarios de Numpy chamam isso de vetorizacao. Qualquer operacao artitimetica entre arrays de mesmo tamanho e aplicada a todos os elementos

```python
arr = np.array([[1., 2., 3.], [4., 5., 6.]])
```

```python
print(arr)
```

## 1.4.1 Multiplicacao

```python
print(arr * arr)
```

## 1.4.2 subtracao

```python
print(arr - arr)
```

## 1.4.2 soma

```python
print(arr + arr)
```

## 1.4.2 divisao

```python
print(arr / arr)
```

## 1.4.3 Operacaoes aritimeticas escalares

```python
print(1 / arr)
```

```python
print(arr ** 2)
```

## 1.4.3 comparacoes entre arrays

```python
arr2 = np.array(([[0., 4., 1.], [7., 2., 12.]]))
```

```python
print(arr2)
```

```python
print(arr2 > arr)
```
