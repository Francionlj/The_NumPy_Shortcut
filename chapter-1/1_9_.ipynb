# 1.6 Indexacao sofisticada

```python
import numpy as np
```

Indexacao sofisticada (fancy index) e um termo adotado pelo numpy para descrever a indexacao como o uso de arrays de interiros.  Suponha que tivemos um array 8x4:

```python
arr = np.zeros((8, 4))
```

```python
print(arr)
```

```python
for i in range(8):
    arr[i] = i
```

```python
print(arr)
```

Para selecionar um subconjunto das linhas em uma ordem especifica,  basta passar uma lista ou um ndarray de inteiros especificos a ordem desejada

```python
print(arr[[4, 3, 0, 6]])
```

Usar indices negativos seleciona as linhas a partir do final

```python
print(arr[[-3, -5, -7]])
```

Passar arrays com varios indices produz um resultado um pouco diferente; e selecionado um array unidimencional de elementos correspondentes a cada tupla de indices

```python
arr = np.arange(32).reshape((8, 4))
```

```python
print(arr)
```

```python
arr[[1, 5, 7, 2], [0, 3, 1, 2]]
```

Para saber mais sobre a funcao resahpe consulte a pasata a parte de Numpy avancado desse repositorio.

Aqui os elementos (1, 0), (5, 3), (7, 1) e (2,2) foram selecionados. Os resultados indexacao sofisticada com um numero de arrays de inteiros equivalenete ao numero de eixo e sempre unidimencional.

O comportamento da indexacao sofisticada nesse caso e um pouco diferente do que alguns usuarios esperam, que e a regiao retangular formada pela selecao de um subconjunto das linhas e colunas da matriz.  seja siquer uma maneira de obter isso:


```python
print(arr[[1, 5, 7, 2]][:, [0, 3, 1, 2]])
```

Lembrece que a indexacao sofisticada ao contrario, ao da fatiamento, sempre copia os dados em um novo array ao atribuir o resultado a uma nova variavel. se voce atribui valores com a indexacao sofisticada,  os valores com a indexacao sofisticada os valores indexados serao modificados:

```python
arr[[1, 5, 7, 2], [0, 3, 1, 2]]
```

```python
arr[[1, 5, 7, 2], [0, 3, 1, 2]] = 0
```

```python
print(arr)
```
