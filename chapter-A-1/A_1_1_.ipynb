# 7.1 Simulacao de muitos passeios aleatorios ao mesmo tempo

```python
import numpy as np
from matplotlib.pyplot import hist2d
```

Se o que voce desaja e simular muitos passeios aleatoriso,  digamos cinco mil pode gerar todos eles fazendo poucas modificacoes no codigo anterior. Se receberem uma tupla de dois elementosm as fincoes de `numpy.random` gerarao um array bidimencional de arremessos de moeda e poderemos calcular a soma cumulativa para cada linha e tambem todos as 5000 passos aleatorios de uma vez so

```python
rng = np.random.default_rng(123445)
```

```python
nwalk = 5_000
```

```python
nsteps = 1_000
```

```python
draws = rng.integers(0, 2, size=(nwalk,  nsteps))
```

```python
steps = np.where(draws > 0, 1, -1)
```

```python
walks = steps. cumsum(axis = 1)
```

```python
print(walks)
```

```python
walks.max()
```

```python
walks.min()
```

Vamos calcular o tempo minimo de travessia para a chegada a 30 ou -30 passos. Isso e um pouco complicado porque nem todos os 5.000 passeios alcancam 30. podemos fazer a verificacao usando ometodo any:

```python
hits30 = (np.abs(walks) >= 30).any(axis = 1)
```

```python
hits30
```

```python
hits30.sum()
```

POdemos usar esse array booleano para selecionar as linhas de passeios (walks) que realmente cruzaram o nivel absoluto de 30 e chamamos `argmax` no eixo 1 para obter os tempos travessia:

```python
crosing_times = (np.abs(walks[hits30])>= 30).argmax(axis = 1)
```

```python
crosing_times
```

Por fim vamos calcular a media dos tempos de travessia

```python
crosing_times.mean()
```
