# 5.1 Entrada e saida de qarquivos com arrays

```python
import numpy as np
```

O Numpy pode salvar e carregar dados de e para discos em algumas formas de texto binario. Neste sacao discutirei apenas o formato binario interno do Numpy, ja que a maioria dos usuarios prefere o pandas e outras ferramentas para carregamento e formatos de arquivos para ver mais informacoes

`numpy.save`,  `numpy,load` sao duas principais funcoes de trabalho para salvamento e o carregamento eficiente de dados de arrays em disco. Os arrays sao salvos por padrao em um formato binario bruto descopactado com estacap de arquivo `. npy`

```python
arr = np.arange(10)
```

```python
arr
```

```python
np.save("some_array", arr)
```

Se o caminho do arquivo nao terminar com `.npy`  a extencao sera acrescida. o array armazena no disco podera entao carregado com o `np.load`

```python
np.load("some_array.npy")
```

Voce pode salvar varios arrays em um arquivo descompactado usando `numpy.savez` e passando os arrays como arsumento nomeados

```python
np.savez("array_archive.npz", a = arr,  b = arr)
```

```python
arch = np.load('array_archive.npz')
```

```python
arch["b"]
```

Se os dados apresentados uma voa compactacao pode ser melhor usar `numpy.save_compressed()`

```python
np.savez_compressed("array_archive.npz", a = arr,  b = arr)
```
