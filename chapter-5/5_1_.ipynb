# 4.6 Unidade e outras logicas de conjuntos

```python
import numpy as np
```

O Numpy tem algumas operacoes basicas de conjuntos para ndarrays unidimencionais. Uma que e muito usado e `numpy. unique`,  que retorna os valres unicos ordenados de um array:

```python
names = np.array(['Bob',  'Will',  'Joe',  'Bob', 'Will',  'Joe',  'Joe'])
```

```python
names
```

```python
np.unique(names)
```

```python
ints = np.array([3, 3, 3, 2, 2, 1, 1, 4, 4])
```

```python
np.unique(ints)
```

Compare `numpy.unique` com a alternativas em Python puro:


```python
sorted(set(names))
```

Em muitos casos a versao do Numpy e mais rapido, e retorma um array Numpy em vez de uma lista Python

Outra funcao ,  `numpy.in1d` verifica se os valores de uma rray sao equivalentes ou de outro array,  retornando um arrau booleano:

```python
values = np.array([6, 0, 0, 3, 2, 5, 6])
```

```python
np.isin(values, [2, 3, 6])
```

```python
np.unique(values)
```

```python
np.intersect1d(values, [2, 3, 6])
```

```python
np.union1d(values, [2, 3, 6])
```

```python
np.isin(values, [2, 3, 6])
```

```python
np.setdiff1d(values, [2, 3, 6])
```

```python
np.setxor1d(values, [2, 3, 6])
```
