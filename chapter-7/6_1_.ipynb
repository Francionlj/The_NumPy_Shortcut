As operacoes de algebra linear com a multiplicacao de matrizes,  decomposicao,  determinante,  e outras operacoes matematicas em matrizes quadradas,  sao uma parte importante de muitas biblioptecas de arrays.  A multiplicacao de dois arrays bidimencionais `*` corresponde ao produto de todos os elementos, enquanto as multiplicacoe de matrizes requerem o uso de uma funcao.  logo existe uma funcao `dor` que e tanto um metodo de array quanto uma funcao do namespace `numoy` para a multiplicacao de matiz

```python
x = np.array([[1., 2., 3.], [4., 5., 6.]])
```

```python
y = np.array([[6., 23.], [-1, 7], [8, 9]])
```

```python
print(x)
```

```python
print(y)
```

```python
print(x.dot(y))
```

`x.dot(y)` e equivalente ao `np.dot(x,  y)`

```python
print(np.dot(x,  y))
```

Um produto de matrizes entre um array bidimencional e um array undimencional de inverso e detreminante

```python
from numpy.linalg import inv,  qr
```

```python
rng = np.random.default_rng(12345)
```

```python
X = rng.standard_normal((5, 5))
```

```python
mat = X.T @ X
```

```python
print(mat)
```

```python
print(inv(mat))
```

```python
print(mat @ inv(mat))
```
