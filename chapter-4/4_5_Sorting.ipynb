## 4.5.1 In-place sorting with `.sort()`

You can sort an array in-place with the `sort` method.

```
rng = np.random.default_rng(seed=12345)
arr = rng.standard_normal(6)
print(f"Original array: {arr}")
```

```
arr.sort()
print(f"Sorted array: {arr}")
```

## 4.5.2 Sorting along an axis

You can sort each one-dimensional section of values in a multidimensional array in-place along an axis by passing the axis number to `sort`.

```
arr = rng.standard_normal((5, 3))
print(f"Original array:\n{arr}")
```

```
# Sort along axis 0 (down the columns)
arr.sort(axis=0)
print(f"Sorted along axis 0:\n{arr}")
```

```
# Sort along axis 1 (across the rows)
arr.sort(axis=1)
print(f"Sorted along axis 1:\n{arr}")
```

## 4.5.3 Sorting with `np.sort()`

The top-level method `np.sort` returns a sorted **copy** of an array instead of modifying the array in-place.

```
arr2 = np.array([5, -10, 7, 1, 0, -3])
print(f"Original array: {arr2}")
```

```
sorted_arr2 = np.sort(arr2)
print(f"Sorted copy: {sorted_arr2}")
print(f"Original array remains unchanged: {arr2}")
```
