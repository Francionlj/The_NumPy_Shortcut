# 4.4 Metodos de arras booleanos

```python
import numpy as np
```

Os valores booleanos estao sujeitos a coercao para 1 `True` e 0 `False` nos metodos anteriores. Logo com frequencia sum e usado com meio de contra valores `True` em um array boleano

```python
rng = np.random.default_rng(12345)
```

```python
arr = rng.standard_normal(100)
```

```python
arr
```

```python
(arr > 0).sum()
```

```python
(arr <= 0).sum()
```
