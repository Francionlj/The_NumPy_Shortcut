Usar arrays Numpy permite muitos tipod de tarefas processamentos de dados com expressao concisas com arrays, o que de outra forma demandaria a criacao de loop. essas pretica de substituir loops explicitos por expressoes de array e chamado por alguns de vetorisacao. Em geral as oparacoes vetorisadoas com arrays sao suiguinificamente mais rapidas que seu sau equivalencia no python puro, agetando com maior inpacto qualquer tipo de computacao numerica. Posteriomente mo `Numpy avancado` explicarei o broadcastingum metodo poderose para a computacao vom vetorisacao.

Suponha que queremos avaliar `sqrt(xˆ2 + yˆ2)` para uma grande regular vetores.  A funcao `numpy.meshgrid` recebe dois arrays unidimencionais e produz marrizes bidimencionais correspondentes a todos os pares `(x,  y)` dos dois arrays

```python
points = np.arange(-5, 5, 0.01)
```

```python
points
```

```python
xs, ys = np.meshgrid(points, points)
```

```python
print(ys)
```

Para avaliar a funcao basta somente excrever a mesma expressao que escreveria com dois pontops

```python
z = np.sqrt(xs ** 2 + ys ** 2)
```

```python
print(z)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(z,  cmap = 'gray',  extent = [-5, 5, -5, 5])
plt.colorbar()
plt.title("Imagem plot de $\\sqrt{x^2 + y^2}$ para o gradiente de valores")
plt.show()
plt.show()
```

O termo vetorisacao e usado para descrever alguns outros conceitos da Ciencia da computacao, mas estu usando para descrever operacoes com arrays de dados de tipo inteiro executaos de uma so vez em ver de ser preciso percorrer valores usado um looop `for` do Python
