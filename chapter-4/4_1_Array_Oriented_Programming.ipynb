# 4.1 Array-Oriented Programming

```python
import numpy as np
import matplotlib.pyplot as plt
```

Using NumPy arrays enables you to express many kinds of data processing tasks as concise array expressions that might otherwise require writing loops. This practice of replacing explicit loops with array expressions is referred to by some as **vectorization**. In general, vectorized array operations will often be one or two (or more) orders of magnitude faster than their pure Python equivalents.

As a simple example, suppose we wished to evaluate the function `sqrt(x^2 + y^2)` across a regular grid of values.

## 4.1.1 `np.meshgrid`

The `np.meshgrid` function takes two 1D arrays and produces two 2D matrices corresponding to all pairs of `(x, y)` in the two arrays.

```python
# Create a grid of points
points = np.arange(-5, 5, 0.01)
xs, ys = np.meshgrid(points, points)
```

```python
# The `ys` matrix
print(f"ys shape: {ys.shape}")
print(f"ys:\n{ys}")
```

```python
# The `xs` matrix
print(f"xs shape: {xs.shape}")
print(f"xs:\n{xs}")
```

## 4.1.2 Evaluating a function on a grid

Now, evaluating the function is a simple matter of writing the same expression you would write with two points.

```python
z = np.sqrt(xs ** 2 + ys ** 2)
print(f"z shape: {z.shape}")
print(f"z:\n{z}")
```

## 4.1.3 Visualizing the result

We can use `matplotlib` to create an image plot of this 2D array.

```python
plt.imshow(z, cmap='gray', extent=[-5, 5, -5, 5])
plt.colorbar()
plt.title("Image plot of $\\sqrt{x^2 + y^2}$ for a grid of values")
plt.show()
```

## 4.1.4 A Note on Vectorization

The term "vectorization" is used in computer science to describe a variety of concepts. In this context, we use it to describe performing array operations on a whole set of values at once, instead of iterating through them one by one in a Python `for` loop. This is the core of array-oriented programming.
